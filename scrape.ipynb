```python
import requests
from bs4 import BeautifulSoup
import os
import shutil
import time
# The URL of the page

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

save_dir = './konteinerovozy/'

for i in range(1,6):
    url = f'https://autoline.kg/-/konteynerovozy--c34?page={i}'
    # Send a GET request to the page
    response = requests.get(url, headers=headers)
    time.sleep(5)
    # Parse the content of the page with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all image tags
    image_tags = soup.find_all('img')

    # Directory where images will be saved

    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    contaners = soup.find_all('div', class_ = 'description-container')
    a_tags = soup.find_all('a', class_='photo-image-outer')

    hrefs = [a['href'] for a in a_tags if 'href' in a.attrs]
    for link in hrefs:
        
        response = requests.get(link)
        time.sleep(2)

        if response.status_code == 200:
            # Parse the content of the page
            soup_page = BeautifulSoup(response.text, 'html.parser')
            
            # Find all image tags in the parsed page
            img_tags = soup_page.find_all('img', attrs={"data-group": "sales-full-thumb"})
            links_in_page = [a['data-src'] for a in img_tags if 'data-src' in a.attrs]
            for i in links_in_page:
                    modified_url = i.replace('common', 'big')
                    print(modified_url)
                    
                    img_response = requests.get(modified_url, stream=True)

                    if img_response.status_code == 200:
                            # Construct a safe filename by taking the last part of the URL
                            filename = os.path.basename(modified_url)
                            # Limit the filename length to avoid errors on filesystems with length limits
                            safe_filename = filename.split('/')[-1]
                            file_path = os.path.join(save_dir, safe_filename)
                            
                            # Open a file to write the image data
                            with open(file_path, 'wb') as file:
                                    for chunk in img_response.iter_content(1024):
                                            file.write(chunk)
                            print(f'Downloaded image {i} from {modified_url}')
                    else:
                            print(f'Failed to download image {i} from {modified_url}')

            print('Done!')
```
